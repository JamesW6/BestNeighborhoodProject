## Median income in each neighborhood vs Park Space

```python
%matplotlib inline
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fileLoc=os.path.join("..", "datasets","neighborhoodData.csv")
df = pd.read_csv(fileLoc)
df["Median Home Value ($ Ten Thousands)"]=df["Median Home Value ($ Ten Thousands)"].replace('[\$,]', '', regex=True).astype(float)
df["Median Home Value ($ Ten Thousands)"]=df["Median Home Value ($ Ten Thousands)"].divide(10000)
df=df.sort_values(by="Median Home Value ($ Ten Thousands)")
df=df.nlargest(columns="Median Home Value ($ Ten Thousands)",n=20)
df.plot(
    kind="bar",
    x="Neighborhood_2010_HOOD",
    y=["Median Home Value ($ Ten Thousands)", "Park Space Acres Per 1000 People"], 
    title="Median Home Value by Neighborhood",
    figsize=(16,6)
)
plt.xlabel("Neighborhoods")
plt.ylabel("Median Home Value, Park Space")
plt.tight_layout()
plt.show()
```

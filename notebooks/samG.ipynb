```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
poverty = pd.read_csv("../datasets/poverty-status-of-individuals-in-the-past-12-months-by-living-arrangement.csv")
povertyRate = poverty["Estimate; Income in the past 12 months below poverty level:"]/poverty["Estimate; Total:"]
neighborhoods = poverty["Neighborhood"]
povertyRatePerNeighborhood = {'Neighborhoods': neighborhoods,'Poverty Rate (%)': povertyRate}
df = pd.DataFrame(povertyRatePerNeighborhood)
df = df.sort_values("Poverty Rate (%)")
df = df.nsmallest(n=20, columns = "Poverty Rate (%)")
df["Poverty Rate (%)"] = df["Poverty Rate (%)"]*100
plt.style.use('classic')
df.plot(kind="bar",
        x="Neighborhoods",
        y="Poverty Rate (%)", 
        title="Poverty Rate by Neighborhood",
        figsize=(15,5), 
        rot = 90,
        color = 'orange')
plt.xlabel("Neighborhoods")
plt.ylabel("Poverty Rate (%)")
plt.show()
```

Using solely this data/submetric, we can conclude that Shouth Shore is the best neighborhood in Pittsburgh. Poverty in a city can lead to greater crime and an overall lower quality of life; a city where there are no impoverished families should correlate to less crime, and an overall happier environment.

# Playgrounds

## Dataset Selection

I chose to use [this](https://data.wprdc.org/dataset/playgrounds) dataset on playgrounds in the city of Pittsburgh.

Public playgrounds have proven to be important for child development, allowing them to socialize in a safe space other than at school. They increase physical activity, and have positive effects on social and mental health. See: https://pmc.ncbi.nlm.nih.gov/articles/PMC11232259/

Playgrounds also provide spaces for parents to socialize and alliviate some of the stress of watching their children, improving their quality of life as well.

## Neighborhood Ranking

Neighborhoods will be ranked by number of playgrounds present. Ties will be allowed.

```python
import pandas
import numpy
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
playgrounds = pandas.read_csv("pitt-playgrounds.csv")
playgrounds["neighborhood"].value_counts().plot.bar(
    figsize=(12, 6),
    title="Playgrounds by Neighborhood",
    xlabel="Neighborhood",
    ylabel="Quantity",
    color=["lightgreen", "green"],
    rot=90
)
plt.show()
```

## Determination

From our data, we can see that the "best" neighborhood according to this particular sub-metric is Squirrel Hill South, with 8 playgrounds.
Beechview and South Side Slopes are tied for second with 5 playgrounds each.
